```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
# from nltk.stem import PorterStemmer
from nltk.tokenize import TweetTokenizer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.sentiment import SentimentIntensityAnalyzer
import csv

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()


# Download NLTK resources (run this once)
nltk.download('stopwords')

# Define Indonesian stop words
stop_words = set(stopwords.words('indonesian'))


def clean(tweet):

    # Remove Twitter handles
    tweet = re.sub(r'@\w+', '', tweet)

    # hilangan hastag
    tweet = re.sub(r'#\w+\s*', '', tweet)

    # Remove URLs
    tweet = re.sub(r'http\S+', '', tweet)

    # Remove special characters and punctuations
    tweet = re.sub(r'[^\w\s]', '', tweet)

    # hilangan special char
    tweet = re.sub(r'[^a-zA-Z0-9\s]', '', tweet)


    return tweet.lower()

def stem_stopWords(tweet):
    # Tokenization
    tokenizer = TweetTokenizer(preserve_case=False, reduce_len=True)
    tokens = tokenizer.tokenize(tweet)

    # Remove stop words
    tokens = [token for token in tokens if token.lower() not in stop_words]

    # # Stemming
    tokens = [stemmer.stem(token) for token in tokens]

    # Join tokens back to form cleaned tweet
    return ' '.join(tokens)

    # return tokens
```

```python
dfTweets = pd.read_csv("tweetsPemilu.csv")
dfTweets = dfTweets.drop_duplicates(subset = "Tweet")
```

```python
dfTweets['clean_tweet'] = dfTweets['Tweet'].apply(clean)
dfTweets
```

```python
dfTweets['stem_rmSW'] = dfTweets['clean_tweet'].apply(stem_stopWords)
dfTweets
```

```python
from transformers import pipeline
```

```python
pretrained_name = "w11wo/indonesian-roberta-base-sentiment-classifier"

nlp = pipeline(
    "sentiment-analysis",
    model=pretrained_name,
    tokenizer=pretrained_name
)
```

```python
def tf_sentiment(tweet):
  return nlp(tweet)[0]['label']
```

```python
dfTweets['tf_sentiment'] = dfTweets['clean_tweet'].apply(tf_sentiment)
dfTweets
```

```python
posDict = {}
with open("/content/InSet/positive.tsv", 'r', newline='') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    next(reader)
    for row in reader:
        posDict[row[0]] = float(row[1])

negDict = {}
with open("/content/InSet/negative.tsv", 'r', newline='') as tsvfile:
    reader = csv.reader(tsvfile, delimiter='\t')
    next(reader)
    for row in reader:
        negDict[row[0]] = float(row[1])
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
indDict = {**posDict, **negDict}
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
sia.lexicon.update(indDict)
```

```python
def lx_sentiment(text):
    comp = sia.polarity_scores(text)['compound']
    if comp<0:
      return 'negatif'
    elif comp>0:
      return 'positif'
    else:
      return 'neutral'
```

```python
dfTweets['lx_sentiment'] = dfTweets['stem_rmSW'].apply(lx_sentiment)
dfTweets
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

pretrained= "mdhugol/indonesia-bert-sentiment-classification"

model = AutoModelForSequenceClassification.from_pretrained(pretrained)
tokenizer = AutoTokenizer.from_pretrained(pretrained)

sentiment_analysis = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

label_index = {'LABEL_0': 'positive', 'LABEL_1': 'neutral', 'LABEL_2': 'negative'}
```

```python
def tf_sentiment2(tweet):
  result = sentiment_analysis(tweet)
  return label_index[result[0]['label']]
  # score = result[0]['score']
```

```python
dfTweets['tf_sentiment2'] = dfTweets['clean_tweet'].apply(tf_sentiment2)
dfTweets
```
