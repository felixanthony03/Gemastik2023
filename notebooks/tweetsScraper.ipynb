```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
query = "pemilu until:2023-08-06 since:2022-01-01"
tweets = []
limit = 5000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():

    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content])
```

```python
len(tweets)
```

```python
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet'])
df
```

```python
df.to_csv('tweetsPemilu.csv', encoding='utf-8', index=False)
```
